## data cleansing : 

```python
import pandas as pd
```

```python
df=pd.read_excel('brandshare.xlsx', skiprows=5)
df
```

```python
df.info
```

```python
dt=df[(df['Geography'] =='Thailand')|(df['Geography'] == 'Japan')]  
dt
```

```python
dt.info()
```

```python
dt['2015']=pd.to_numeric(dt['2015'],errors='coerce')
dt['2016']=pd.to_numeric(dt['2016'],errors='coerce')
dt['2017']=pd.to_numeric(dt['2017'],errors='coerce')
dt['2018']=pd.to_numeric(dt['2018'],errors='coerce')
dt['2019']=pd.to_numeric(dt['2019'],errors='coerce')
dt['2020']=pd.to_numeric(dt['2020'],errors='coerce')
```

```python
dt
```

```python
dt=dt.fillna(value=0)
```

```python
dt
```

```python
dt.rename(columns={'Company Name':'Company_Name'}, inplace=True)
dt
```

```python
dt['2015']=dt['2015'].astype(int)
dt['2016']=dt['2016'].astype(int)
dt['2017']=dt['2017'].astype(int)
dt['2018']=dt['2018'].astype(int)
dt['2019']=dt['2019'].astype(int)
dt['2020']=dt['2020'].astype(int)
```

```python
dt.drop(index=409,inplace=True)
dt.drop(index=632,inplace=True)
```

```python
dt.drop(['Data Type'], axis=1, inplace=True)
dt
```

```python
dt.drop(['Category'], axis=1, inplace=True)
dt
```

```python
pp=dt[dt['Geography'].str.contains(r'Japan', case = False)]
pp
```

```python
jp=dt[dt['Geography'].str.contains(r'Japan', case = False)]
jp
```

```python
jp['2015']=jp['2015']/jp['2015'].sum()*100
jp['2016']=jp['2016']/jp['2016'].sum()*100
jp['2017']=jp['2017']/jp['2017'].sum()*100
jp['2018']=jp['2018']/jp['2018'].sum()*100
jp['2019']=jp['2019']/jp['2019'].sum()*100
jp['2020']=jp['2020']/jp['2020'].sum()*100
jp
```

```python
pt=dt[dt['Geography'].str.contains(r'Thailand', case = False)]
pt
```

```python
th=dt[dt['Geography'].str.contains(r'Thailand', case = False)]
th
```

```python
th['2015']=th['2015']/th['2015'].sum()*100
th['2016']=th['2016']/th['2016'].sum()*100
th['2017']=th['2017']/th['2017'].sum()*100
th['2018']=th['2018']/th['2018'].sum()*100
th['2019']=th['2019']/th['2019'].sum()*100
th['2020']=th['2020']/th['2020'].sum()*100
th
```

```python
th['2015'] = th['2015'].apply(lambda x: float("{:.2f}".format(x)))
th['2016'] = th['2016'].apply(lambda x: float("{:.2f}".format(x)))
th['2017'] = th['2017'].apply(lambda x: float("{:.2f}".format(x)))
th['2018'] = th['2018'].apply(lambda x: float("{:.2f}".format(x)))
th['2019'] = th['2019'].apply(lambda x: float("{:.2f}".format(x)))
th['2020'] = th['2020'].apply(lambda x: float("{:.2f}".format(x)))
th
```

```python
jp['2015'] = jp['2015'].apply(lambda x: float("{:.2f}".format(x)))
jp['2016'] = jp['2016'].apply(lambda x: float("{:.2f}".format(x)))
jp['2017'] = jp['2017'].apply(lambda x: float("{:.2f}".format(x)))
jp['2018'] = jp['2018'].apply(lambda x: float("{:.2f}".format(x)))
jp['2019'] = jp['2019'].apply(lambda x: float("{:.2f}".format(x)))
jp['2020'] = jp['2020'].apply(lambda x: float("{:.2f}".format(x)))
jp
```
